```python
# Reading the dataset
import numpy as np

data_file='/makeSenseOfCensus.csv'
```

```python
data = np.genfromtxt(data_file, delimiter=",", skip_header=1)

# printing the data

print("\nData: \n\n", data)
```

```python
#printing the type of data

print("\nType fo data: \n\n", type(data))
```

```python
# append new record to data using np.concatenate

new_record = [[50,9,4,1,0,0,40,0]]
```

```python
np.concatenate([new_record,data], axis = 0)
```

```python
print("\nData: \n\n", data)
```

```python
# age of people in index 0
age = data[:,0]
print(age)
```

```python
# find max age 
max_age = age.max()
print(max_age)
```

```python
#find min age
min_age = age.min()
print(min_age)
```

```python
#find mean age
age_mean = age.mean()
print(age_mean)
```

```python
#find standard deviation
age_std = age.std()
print(age_std)
```

```python
#number of citizen per race
race_0 = data[data[:,2]==0]
print(race_0)
```

```python
race_1 = data[data[:,2]==1]
race_2 = data[data[:,2]==2]
race_3 = data[data[:,2]==3]
race_4 = data[data[:,2]==4]
print(race_1)
print(race_2)
print(race_3)
print(race_4)
```

```python
#len per race
len_0 = len(race_0)
print('Length of race_0',len_0)
len_1 = len(race_1)
print('Length of race_1',len_1)
len_2 = len(race_2)
print('Length of race_2',len_2)
len_3 = len(race_3)
print('Length of race_3',len_3)
len_4 = len(race_4)
print('Length of race_4',len_4)
```

```python
#find the minimum number of citizen in a race
min_cit = min(len_0,len_1,len_2,len_3,len_4)
```

```python
if min_cit == len_0:
  print('Race_0 has Minimum no. of Citezen of {}.'.format(len_0)) 
elif min_cit == len_1:
  print('Race_1 has Minimum no. of Citezen of {}.'.format(len_1))
elif min_cit == len_2:
  print('Race_2 has Minimum no. of Citezen of {}.'.format(len_2))
elif min_cit == len_3:
  print('Race_3 has Minimum no. of Citezen of {}.'.format(len_3))  
elif min_cit == len_4:
  print('Race_4 has Minimum no. of Citezen of {}.'.format(len_4)) 
else:
  pass
```

```python
#find the number of citizen over age 60
senior_citizen = data[age>60]
```

```python
#find the working hour sum of citizen over age 60
working_hour_sum =sum(senior_citizen[:,6])
print(working_hour_sum)
```

```python
#find len of senior citizen
senior_citizen_len = len(senior_citizen)
```

```python
#find the average working hour of senior citizen
avg_working_hours = working_hour_sum/senior_citizen_len
print(avg_working_hours)
```
